#### Set Constants and paramaters

```python
#input params
H = 0.42 #no units,ratio #Hematocrit assumed to be constant
M = 30* 10e-9 #mol_ O2/(mL/s) #Taken from table 2 from Wiley Payne paper
cHb = 0.2 #mL_O2/m #Taken from table 2 from Wiley Payne paper
paO2_bar_t = 15 #mmHG #Taken from table 2 from Wiley Payne paper
K = 5*10e-8

##reset df
df = None

#Input data from paper
df = pd.DataFrame({'Name': ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'C', 'V6', 'V5', 'V4', 'V3', 'V2', 'V1'], 
                   'Number': [1, 2, 4, 8, 16, 32, 64, 32, 16, 8, 4, 2, 1],
                   'Diameter(µm)': [23.97, 19.17, 15.28, 12.08, 9.46, 7.32, 8, 11.51, 14.53, 17.79, 21.45, 25.70, 30.77],
                   'Length(µm)': [1267.6, 930.3, 543.6, 302.3, 161.2, 154.7, 243.9, 473.9, 272.3, 426.6, 632.5, 844.2, 936.3],
                   'Viscocity(mPAs)': [1.59, 1.50, 1.42, 1.34, 1.28, 1.23, 1.24, 1.33, 1.40, 1.48, 1.55, 1.62, 1.7],
                   'Pressure Drop(mmHg)': [6.93, 5.87, 4.02, 2.70, 1.82, 2.35, 2.62, 1.27, 0.61, 0.89, 1.31, 1.78, 2.01],
                   'Saturation': [0.94, 0.93, 0.92, 0.89, 0.84, 0.765, 0.665, 0.61, 0.5975, 0.5875, 0.5825, 0.5775, 0.5725],
                   'dS': [0.01, 0.01, 0.02, 0.04, 0.0625,0.0875, 0.0775, 0.03375, 0.01125, 0.0075, 0.005, 0.005, 0.005]
                  })
```

#### Calculate Flows and Vt

```python
#Calculate resistances of each vessel
df['Resistance for U'] = 32 * df['Viscocity(mPAs)'] * df['Length(µm)'].div(df['Diameter(µm)']**2)
df['Resistance for Q'] = 128 * df['Viscocity(mPAs)'] * df['Length(µm)'].div(math.pi*(df['Diameter(µm)']**4))

#Alter resistances for C
#These are aross each pressure drop (explains the doubled resistance for the capillaries as in paralell so have half the flow each (flow is shared))
df.loc[df.index[df['Name'] == 'C'][0],'Resistance for U'] = df.loc[df.index[df['Name'] == 'C'][0],'Resistance for U']*2
df.loc[df.index[df['Name'] == 'C'][0],'Resistance for Q'] = df.loc[df.index[df['Name'] == 'C'][0],'Resistance for Q']*2

#Calculate flow and velocity in a single vessel
df['U in single(m/s????)'] = df['Pressure Drop(mmHg)'].div(df['Resistance for U'])
df['Q in single(m^3/s????)'] = df['Pressure Drop(mmHg)'].div(df['Resistance for Q'])

#Find change in saturdation from data on saturation curve
# S = pd.Series([])

#Calculate Vt CHANGE IN SATURATIO  YA SHIT HEAD
df['Vt'] = df['Q in single(m^3/s????)']*cHb*H*df['dS']/M
```

#### Calculate partial pressure in blood
S and partial O2 relationship given by servinghaus 1979  
Checked partial pressure against typical values online and it seems bang on.

```python
def check_imag_roots_real(all_roots):
    counter = 0
    sols = 0
    for i in range(len(all_roots)):
        if all_roots[i].imag <= 10e-13 and all_roots[i].imag >= -10e-13:
            sols = all_roots[i].real
            counter += 1
    if counter != 1:
        print('Something seems wrong with the partial presure to Saturation cubic solver as not one solution is returned.')
    return sols
 
for i in range(len(df['Saturation'])):
    root_temp = None
    sols = 0
    root_temp = np.roots([1,0,150,23400*df.loc[i,'Saturation']/(df.loc[i,'Saturation']-1)]) #gives mmHg , *133.322
    sols = check_imag_roots_real(root_temp) 
    #sols = sols*133.322 ###Remove this step if want in mmHg
    df.loc[i,'partial pressure blood(mmHg)'] = sols
```

#### Calculate wall thickness
Will take my vessel average partial pressure to be the 0.5(pin + pout)  
All units were in mm so i played around with this equation from the paper  
Equations are taken from Payne paper

```python
#all units were in mm so i played around with this equation from the paper
df['X-Area wall(mm^2?)'] = np.pi * ( 0.16*(df['Diameter(µm)'])**2 + 1.4*(df['Diameter(µm)']) + 14 )
df['wall thickness(µm)'] = ( -(df['Diameter(µm)']) + np.sqrt((df['Diameter(µm)'])**2 + (4*(df['X-Area wall(mm^2?)'])/np.pi)) ) / 2
```

#### Calculate partial pressure in tissue
Equations can be found on the week 1 or 2 notes from class. From steady state version of something on the paper

```python
df['tissue partials(mmHg)'] = df['partial pressure blood(mmHg)'] - M * df['Vt'] * df['wall thickness(µm)'] / (2 * np.pi * K * (df['Diameter(µm)']/2) * df['Length(µm)'])
```

#### Find Resistance of baseline model

```python
baselines = df_base[['Name','Number','Resistance for Q','Q in single(m^3/s????)','Vt','wall thickness(µm)','tissue partials(mmHg)','Diameter(µm)','Length(µm)']].rename(columns={"Q in single(m^3/s????)": "baseline Q", "tissue partials(mmHg)": "baseline tissue partial"})
```

```python
#Find Total Resistance. C refers to capilaary and each numer refers to an added level or resistance.
C_ = baselines.loc[6,'Resistance for Q']/2
C_6 = (C_ + baselines.loc[5,'Resistance for Q'] + baselines.loc[7,'Resistance for Q'])/2
C_65 = (C_6 + baselines.loc[4,'Resistance for Q'] + baselines.loc[8,'Resistance for Q'])/2
C_654 = (C_65 + baselines.loc[3,'Resistance for Q'] + baselines.loc[9,'Resistance for Q'])/2
C_6543 = (C_654 + baselines.loc[2,'Resistance for Q'] + baselines.loc[10,'Resistance for Q'])/2
C_65432 = (C_6543 + baselines.loc[1,'Resistance for Q'] + baselines.loc[11,'Resistance for Q'])/2
C_654321 = C_65432 + baselines.loc[0,'Resistance for Q'] + baselines.loc[12,'Resistance for Q']
R_total = C_654321

#Proves for the baseline conditions that it gives the right total flow from my resitance calcs.
#34.18 is pressure drop
Q_total = 34.18 / C_654321
print(Q_total)
```

```python
saturation_in = 0.945
baseline_pressure_in = 60
baseline_pressure_difference = 34.18
pressure_out = baseline_pressure_in - baseline_pressure_difference #pressure_out is a constant
p_percent_change = 0.5
no_runs = 100

#need R_total?
#########################################################################
def new_run(baselines, P_in, P_out, saturation_in):
   
    pressure_difference = P_in - P_out

    Q_in = pressure_difference/R_total
    run = pd.DataFrame({'Name': ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'C', 'V6', 'V5', 'V4', 'V3',
           'V2', 'V1']})
    
    run['Q'] = np.empty
    run.loc[0,'Q'] = Q_in
    run.loc[1,'Q'] = run.loc[0,'Q']/2
    run.loc[2,'Q'] = run.loc[1,'Q']/2
    run.loc[3,'Q'] = run.loc[2,'Q']/2
    run.loc[4,'Q'] = run.loc[3,'Q']/2
    run.loc[5,'Q'] = run.loc[4,'Q']/2
    run.loc[6,'Q'] = run.loc[5,'Q']/2
    run.loc[7,'Q'] = run.loc[6,'Q']*2
    run.loc[8,'Q'] = run.loc[7,'Q']*2
    run.loc[9,'Q'] = run.loc[8,'Q']*2
    run.loc[10,'Q'] = run.loc[9,'Q']*2
    run.loc[11,'Q'] = run.loc[10,'Q']*2
    run.loc[12,'Q'] = run.loc[11,'Q']*2

    run['dS'] = ( baselines['Vt'] * M ) / ( cHb * H * run['Q'] )
    run['Saturation'] = np.empty

    #due to the 0.945 being from an inital value above.
    run.loc[0,'Saturation'] = saturation_in - 0.5*run.loc[0,'dS']
    run.loc[1,'Saturation'] = run.loc[0,'Saturation'] - 0.5*run.loc[0,'dS'] - 0.5*run.loc[1,'dS']
    run.loc[2,'Saturation'] = run.loc[1,'Saturation'] - 0.5*run.loc[1,'dS'] - 0.5*run.loc[2,'dS']
    run.loc[3,'Saturation'] = run.loc[2,'Saturation'] - 0.5*run.loc[2,'dS'] - 0.5*run.loc[3,'dS']
    run.loc[4,'Saturation'] = run.loc[3,'Saturation'] - 0.5*run.loc[3,'dS'] - 0.5*run.loc[4,'dS']
    run.loc[5,'Saturation'] = run.loc[4,'Saturation'] - 0.5*run.loc[4,'dS'] - 0.5*run.loc[5,'dS']
    run.loc[6,'Saturation'] = run.loc[5,'Saturation'] - 0.5*run.loc[5,'dS'] - 0.5*run.loc[6,'dS']
    run.loc[7,'Saturation'] = run.loc[6,'Saturation'] - 0.5*run.loc[6,'dS'] - 0.5*run.loc[7,'dS']
    run.loc[8,'Saturation'] = run.loc[7,'Saturation'] - 0.5*run.loc[7,'dS'] - 0.5*run.loc[8,'dS']
    run.loc[9,'Saturation'] = run.loc[8,'Saturation'] - 0.5*run.loc[8,'dS'] - 0.5*run.loc[9,'dS']
    run.loc[10,'Saturation'] = run.loc[9,'Saturation'] - 0.5*run.loc[9,'dS'] - 0.5*run.loc[10,'dS']
    run.loc[11,'Saturation'] = run.loc[10,'Saturation'] - 0.5*run.loc[10,'dS'] - 0.5*run.loc[11,'dS']
    run.loc[12,'Saturation'] = run.loc[11,'Saturation'] - 0.5*run.loc[11,'dS'] - 0.5*run.loc[12,'dS']

    for i in range(len(run['Saturation'])):
        root_temp = None
        sols = 0
        root_temp = np.roots([1,0,150,23400*run.loc[i,'Saturation']/(run.loc[i,'Saturation']-1)])
        sols = check_imag_roots_real(root_temp) 
        run.loc[i,'partial pressure blood(mmHg)'] = sols
    
    run['tissue partials(mmHg)'] = run['partial pressure blood(mmHg)'] - M * baselines['Vt'] * baselines['wall thickness(µm)'] / (2 * np.pi * K * (baselines['Diameter(µm)']/2) * baselines['Length(µm)'])

    return run
#################################################
    
run_input_pressures = pd.DataFrame({'Pressure in(mmHg)': np.linspace((baseline_pressure_in-baseline_pressure_difference*p_percent_change),(baseline_pressure_in+baseline_pressure_difference*p_percent_change), num=no_runs), 'Pressure out(mmHg)': np.zeros(no_runs) + pressure_out})

master = {}

for i in range(no_runs):
    P_in = run_input_pressures.loc[i,'Pressure in(mmHg)'] 
    P_out = run_input_pressures.loc[i,'Pressure out(mmHg)']
    master[i] = new_run(baselines, P_in, P_out, saturation_in) #get i to label it properly
```

#### Pull out vessel data
Create a function to pull out the flow and tissue pressure data into a dataframe which seperates them out for each vessel

```python
flow = pd.DataFrame({'A1': np.zeros(len(master)), 'A2': np.zeros(len(master)), 'A3': np.zeros(len(master)), 'A4': np.zeros(len(master)), 'A5': np.zeros(len(master)), 'A6': np.zeros(len(master)), 'C': np.zeros(len(master)), 'V6': np.zeros(len(master)), 'V5': np.zeros(len(master)), 'V4': np.zeros(len(master)), 'V3': np.zeros(len(master)),
           'V2':np.zeros(len(master)), 'V1': np.zeros(len(master))})
tissue_partials = pd.DataFrame({'A1': np.zeros(len(master)), 'A2': np.zeros(len(master)), 'A3': np.zeros(len(master)), 'A4': np.zeros(len(master)), 'A5': np.zeros(len(master)), 'A6': np.zeros(len(master)), 'C': np.zeros(len(master)), 'V6': np.zeros(len(master)), 'V5': np.zeros(len(master)), 'V4': np.zeros(len(master)), 'V3': np.zeros(len(master)),
           'V2':np.zeros(len(master)), 'V1': np.zeros(len(master))})
flow_normal = pd.DataFrame({'A1': np.zeros(len(master)), 'A2': np.zeros(len(master)), 'A3': np.zeros(len(master)), 'A4': np.zeros(len(master)), 'A5': np.zeros(len(master)), 'A6': np.zeros(len(master)), 'C': np.zeros(len(master)), 'V6': np.zeros(len(master)), 'V5': np.zeros(len(master)), 'V4': np.zeros(len(master)), 'V3': np.zeros(len(master)),
           'V2':np.zeros(len(master)), 'V1': np.zeros(len(master))})
tissue_partials_normal = pd.DataFrame({'A1': np.zeros(len(master)), 'A2': np.zeros(len(master)), 'A3': np.zeros(len(master)), 'A4': np.zeros(len(master)), 'A5': np.zeros(len(master)), 'A6': np.zeros(len(master)), 'C': np.zeros(len(master)), 'V6': np.zeros(len(master)), 'V5': np.zeros(len(master)), 'V4': np.zeros(len(master)), 'V3': np.zeros(len(master)),
           'V2':np.zeros(len(master)), 'V1': np.zeros(len(master))})


for i in range(len(master)):
    for j in range(len(baselines)):
        flow.iloc[i,j] = master[i].loc[j,'Q'] 
        tissue_partials.iloc[i,j] = master[i].loc[j,'tissue partials(mmHg)']

for i in range(len(baselines)):
    flow_normal.iloc[:,i] = flow.iloc[:,i] / baselines.loc[i,'baseline Q']
    tissue_partials_normal.iloc[:,i] = tissue_partials.iloc[:,i] / baselines.loc[i,'baseline tissue partial']
```

#### Now plot the values

```python
fig,ax = plt.subplots(figsize=(10,8))

for i in range(len(baselines)):
    plt.plot(flow_normal.iloc[:,i],tissue_partials_normal.iloc[:,i], label=baselines.loc[i,'Name'])

ax.legend()
plt.grid(which='both')
plt.ylabel('normalised tissue partials')
plt.xlabel('flow normalised')
ax.set_ylim(bottom=0.4, top=1.3)
plt.title('Normalised partial tissue pressure against normalised flow in individual vessels', fontsize=20)
plt.savefig('tissue partials against flow')
```
