```python
t = pd.Series(np.linspace(0,max_time,no))
dt = max_time/(no-1)
drop_value = 60 - 34.18*ratio_drop

#### PRessure stuff
pressure_in = pd.Series(np.zeros(len(t)))
for i in range(len(t)):
    if t[i] <= delay_pressure_drop:
        pressure_in[i] = 60
    elif t[i] <= delay_pressure_drop + time_for_drop:
        pressure_in[i] = 60 - ((t[i] - delay_pressure_drop)/time_for_drop)*(60-drop_value)
    else:
        pressure_in[i] = drop_value
pressure_out = pd.Series(np.zeros(len(t))) + 60 - 34.18
pressure_difference = pressure_in - pressure_out

print('Time of simulation :',max_time,'seconds')
print('Time step :',dt,'seconds')
print('Number of iterations:',no)
```

## Model

#### Functions

```python
def total_R(baselines,phi,alpha):
    C_ = alpha*(baselines.at[6,'Resistance for Q']/2)/phi**4  + (1-alpha)*(baselines.at[6,'Resistance for Q']/2)
    C_6 = (C_ + baselines.at[5,'Resistance for Q'] + baselines.at[7,'Resistance for Q'])/2
    C_65 = (C_6 + baselines.at[4,'Resistance for Q'] + baselines.at[8,'Resistance for Q'])/2
    C_654 = (C_65 + baselines.at[3,'Resistance for Q'] + baselines.at[9,'Resistance for Q'])/2
    C_6543 = (C_654 + baselines.at[2,'Resistance for Q'] + baselines.at[10,'Resistance for Q'])/2
    C_65432 = (C_6543 + baselines.at[1,'Resistance for Q'] + baselines.at[11,'Resistance for Q'])/2
    C_654321 = C_65432 + baselines.at[0,'Resistance for Q'] + baselines.at[12,'Resistance for Q']
    R_total = C_654321
    return R_total
 
def euler(alpha,delay,baselines):
    delay_number = round(delay/dt)
    
    R_tot = pd.Series(np.zeros(len(t)))
    Q_tot = pd.Series(np.zeros(len(t)))
    Q_norm = pd.Series(np.ones(len(t)))
    
    phi = pd.Series(np.ones(no))
    phi_min = pd.Series(np.ones(no))
    dphidt = pd.Series(np.zeros(no))
    
    for i in tqdm(range(len(t))):
        if i >=delay_number: 
            R_tot[i] = total_R(baselines,phi[i],alpha)
            Q_tot[i] = pressure_difference[i]/R_tot[i]
            Q_norm[i] = Q_tot[i] / baselines['Q in single(µm3/s)'][0]
            dphidt[i] =  (1/optimised_value_tau)*( -phi[i-delay_number] + Q_norm[i]*(1-phi_min[i]) + phi_min[i] )
            phi[i+1] = dphidt[i]*dt + phi[i]
            phi_min[i+1] = 1-(alpha**0.25)*(1-phi_min_baseline)*(1-Q_norm[i])**n
        else:
            ###############################################################################
            #URGENTLY NEED TO FIX THIS BIT UP PROPERLY#
            ###############################################################################
            Q_tot[i] = 3.713608e+06
#             Q_tot[i] = pressure_difference[0]/R_tot[0] # Havent tried putting this into model but dont see why it wouldnt work
            
    combined = pd.DataFrame()
    combined['t'] = t
    combined['pressure_difference'] = pressure_difference
    combined['phi'] = phi
    combined['phi_min'] = phi_min
    combined['dphidt'] = dphidt
    combined['R_tot'] = R_tot
    combined['Q_tot'] = Q_tot
    combined['Q_norm'] = Q_norm
            
    return combined
```

```python
def check_imag_roots_real(all_roots):
    counter = 0
    sols = 0
    for i in range(len(all_roots)):
        if all_roots[i].imag <= 10e-13 and all_roots[i].imag >= -10e-13:
            sols = all_roots[i].real
            counter += 1
    if counter != 1:
        print('Something seems wrong with the partial presure to Saturation cubic solver as not one solution is returned.')
    return sols
```

```python
# Initialise Data, set up DFs
baseline = baselines.drop(['Viscocity(mmHg*s)','Resistance for U', 'U in single(µm/s)'],axis=1)
first_iteration = baseline.loc[:,['Name','Number','Diameter(µm)','Length(µm)','Saturation in','Saturation out','Resistance for Q','Q in single(µm3/s)','Vt(µm3)','partial pressure blood(mmHg)','X-Area wall(???)','wall thickness(µm)','tissue partials(mmHg)']]
constant = baseline.drop(['Q in single(µm3/s)','Pressure Drop(mmHg)','Saturation in','Saturation out','Resistance for Q','dS','Saturation ave','partial pressure blood(mmHg)','X-Area wall(???)','wall thickness(µm)','tissue partials(mmHg)'], axis=1)
vessel_data_alltime = {}
pt_averaged = pd.Series(np.zeros(len(t)))
terms_data_alltime = {}

def run():
    for i in tqdm(range(len(t))):

        if i ==0:
            step = first_iteration.copy()
            step['dptdt'] = 0
        else:       
            step = vessel_data_alltime[i-1].copy()
            step['tissue partials(mmHg)'] = vessel_data_alltime[i-1]['tissue partials(mmHg)'] + vessel_data_alltime[i-1]['dptdt']*dt
            step['Saturation out'] = vessel_data_alltime[i-1]['Saturation out'] + vessel_data_alltime[i-1]['dSoutdt']*dt
            step['Saturation in'] = vessel_data_alltime[i-1]['Saturation in'] + vessel_data_alltime[i-1]['dSindt']*dt

        step.at[6,'Diameter(µm)'] = baseline.at[6,'Diameter(µm)']*out.at[i,'phi']    
        step.at[6,'X-Area wall(???)'] = np.pi * ( 0.16*(step.at[6,'Diameter(µm)'])**2 + 1.4*(step.at[6,'Diameter(µm)']) + 14 )
        step.at[6,'wall thickness(µm)'] = ( -(step.at[6,'Diameter(µm)']) + np.sqrt((step.at[6,'Diameter(µm)'])**2 + (4*(step.at[6,'X-Area wall(???)'])/np.pi))) / 2
        step['Saturation ave'] = (step['Saturation in']+step['Saturation out'])/2

        step['M'] = M_constant*step['tissue partials(mmHg)']/(step['tissue partials(mmHg)']+hill_constant)

        for j in range(len(step['Saturation ave'])):
            root_temp = None
            sols = 0
            root_temp = np.roots([1,0,150,23400*step.at[j,'Saturation ave']/(step.at[j,'Saturation ave']-1)]) #gives mmHg
            sols = check_imag_roots_real(root_temp) 
            step.at[j,'partial pressure blood(mmHg)'] = sols

        step['Q in single(µm3/s)'] = np.empty
        terms = pd.DataFrame({'Name': ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'C', 'V6', 'V5', 'V4', 'V3', 'V2', 'V1']})

        #Could pull the Q stuff all out and do it seperately from this bit.
        for j in range(len(step)):
            if j ==0:
                step.at[j,'dSindt'] = 0
                step.at[j,'Q in single(µm3/s)'] = out.at[i,'Q_tot']
            else:
                step.at[j,'dSindt'] = step.at[j-1,'dSoutdt']
                if j <=  6:
                    step.at[j,'Q in single(µm3/s)'] = step.at[j-1,'Q in single(µm3/s)']/2
                else:
                    step.at[j,'Q in single(µm3/s)'] = step.at[j-1,'Q in single(µm3/s)']*2
                    
            step.at[j,'dSoutdt'] = 2/(np.pi*((step.at[j,'Diameter(µm)']**2)/4)*step.at[j,'Length(µm)']) \
                                    * ( - (2 * np.pi * K * (step.at[j,'Diameter(µm)']/2) * constant.at[j,'Length(µm)'])/(step.at[j,'wall thickness(µm)']*cHb*H) \
                                    * (step.at[j,'partial pressure blood(mmHg)']-step.at[j,'tissue partials(mmHg)'])\
                                    - step.at[j,'Q in single(µm3/s)']*(step.at[j,'Saturation out']-step.at[j,'Saturation in']) )\
                                    - step.at[j,'dSindt']
            
        step['dptdt'] = (1/(alpha_t*constant['Vt(µm3)']))\
                        *(((2 * np.pi * K * (step['Diameter(µm)']/2) * constant['Length(µm)'])/(step['wall thickness(µm)']))\
                        *(step['partial pressure blood(mmHg)']-step['tissue partials(mmHg)'])\
                        - step['M'] * constant['Vt(µm3)'])
        
        step['pt_by_Vt'] = step['tissue partials(mmHg)'] * step['Vt(µm3)']
        pt_averaged[i] = step['pt_by_Vt'].sum()/13

        vessel_data_alltime[i] = step
        
    if i == no-1:
        print('finished')
    else:
        print('broken')

# %timeit -r 10 -n 1 run()
        
# %prun -s cumulative run()

# %lprun -f run run()

run()
```
