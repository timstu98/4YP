```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Rdivk = np.arange(0,10,0.01)
```

```python
ks = [0.05,0.1,0.2,0.3]
for k in ks:
    c_init = Rdivk/(1/k+1)
    name = 'k =' + str(k)
    plt.plot(Rdivk,c_init,label=name)
c_final = Rdivk
plt.plot(Rdivk,c_final)
plt.legend()
```

```python
plt.plot(Rdivk,c_final)
plt.plot(Rdivk,c_init)
```

```python
k = np.arange(0.1,1,0.01)
Rs = [0.1,0.2,0.3,0.4,1]
color = []
for R in Rs:
    c_init = R/(1+k)
    name = 'cinit R =' + str(R)
    plt.plot(k,c_init,label=name)
    c_final = R/k
    name = 'cfinal R =' + str(R)
    plt.plot(k,c_final,label=name)
plt.legend()
```

```python
R = np.arange(0.05,2,0.01)
ks = [0.05,0.1]
for k in ks:
    c_init = R/(1+k)
    name = 'cinit k =' + str(k)
    plt.plot(R,c_init,label=name,linestyle = '--')
    c_final = R/k
    name = 'cfinal k =' + str(k)
    plt.plot(R,c_final,label=name)
plt.legend()
```
